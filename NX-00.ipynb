```python
import numpy
import pandas
import matplotlib.pyplot as plt
import networkx as nx
```

<h2>初始圖購</h2>

```python
G = nx.Graph()
```

<h2>建構圖資</h2>

```python
#G = nx.petersen_graph()
```

```python
G.clear()
```

```python
for i in range(5):
    G.add_node(i)
```

```python
G.add_edge(0,1)#節點一與節點三連在一起
G.add_edge(0,2)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,4)
G.add_edge(2,5)
```

<h2>顯示圖</h2>

```python
fig,(ax1,ax2) = plt.subplots(dpi=240, ncols=2)
nx.draw(G,ax=ax1,with_labels=True,font_weight='bold')
nx.draw_shell(G,ax=ax2,with_labels=True,font_weight='bold')
fig.tight_layout()
plt.show()
```
