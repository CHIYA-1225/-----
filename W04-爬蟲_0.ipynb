```python
import requests #處理http,url
from bs4 import BeautifulSoup #爬蟲套件
import codecs #讀取外部檔案
import json
```

```python
import pandas
from pandas import DataFrame
```

```python
def grab_vendors(page_no=1): #ctrl+U (檢查網頁原始碼)
    url = '%s/manufacturers/index.php?page=%d' % (url_domain, page_no)
    response = requests.get(url) #json
    soup = BeautifulSoup(response.text, 'html.parser')#轉換資料結構
    
    result = soup.find_all('div', {'class': 'Txt b-box'})
    company_list = []
    #過濾讀到的資料
    for ele in result:
        if ele.text.strip() == '':
            continue

        _company_info = []

        for tag in ele.find_all():
            if tag.name == 'h3':
                _company_info.append(ele.text.strip().split('\n'))

        company_list.append(_company_info[0])   
        
    return company_list
```

```python
url_domain = 'https://www.edomtech.com/tw'

company_list = dict()

for page_no in range(1, 7):
    for vendor in grab_vendors(page_no):
        if len(vendor) == 1:
            vendor.append(vendor[0])
            
        company_list[vendor[0]] = vendor[1]
company_list
```

```python
#儲存擷取到的json
with codecs.open('C:\\Users\\user\\anaconda3\\Projects\\BizCases\\data/製造商-清單.json', 'w', encoding='utf8') as W:
    json.dump(company_list, W, sort_keys=True, indent=4, ensure_ascii=False)
```
